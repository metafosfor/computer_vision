```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
# Define the coordinates of the two points

x1 = 1
y1 = 2

x2 = 2
y2 = 3

x3 = 3
y3 = 4

x_coordinates = [x1, x2, x3]
y_coordinates = [y1, y2, y3]

# Create a plot
plt.figure()

# Plot the points
plt.plot(x_coordinates, y_coordinates, 'o', label='Дві точки')
plt.xlim(-10, 10)  # Обмеження осі X від 0 до 5
plt.ylim(-10, 10) # Обмеження осі Y від -2 до 2

plt.axhline(y=0, color='black', linewidth=1.5, linestyle='-', label='Вісь X')
plt.axvline(x=0, color='black', linewidth=1.5, linestyle='-', label='Вісь Y')

# Add labels and a title
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Графік з двома точками')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
m1 = np.linspace(-5, 5, 100) 
c1 = x1 - m1*y1

m2 = np.linspace(-5, 5, 100)
c2 = x2 - m2*y2

m3 = np.linspace(-5, 5, 100)
c3 = x3 - m3*y3

# 3. Будуємо графік, передаючи X першим, а потім Y
plt.plot(m1, c1, label=f'x1 = {x1}y1 + {y1}', color='blue')
# 3. Будуємо графік, передаючи X першим, а потім Y
plt.plot(m2, c2, label=f'x2 = {x2}y2 + {y2}', color='red')
# 3. Будуємо графік, передаючи X першим, а потім Y
plt.plot(m3, c3, label=f'x3 = {x3}y3 + {y3}', color='green')

plt.xlim(-10, 10)  # Обмеження осі X від 0 до 5
plt.ylim(-10, 10) 

# Додаємо осі координат
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)

# Додаємо підписи та заголовок
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Графік з двома лініями')

plt.grid(True)
plt.legend()

# Показуємо графік
plt.show()
```
